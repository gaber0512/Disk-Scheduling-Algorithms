```python
def fcfs(reqs, start):
    if start not in reqs :
        reqs.insert(0,start)
    else:
        reqs.remove(start)
        reqs.insert(0,start)
    return reqs
```

```python

requests=[14,11,16,4,35,40]
def sstf(reqs,start):
    if start not in reqs :
        reqs.append(start)
    x=start
    r=[]
    while reqs:
        r.append(x)
        reqs.sort()

        if len(reqs)>=2:
            i=reqs.index(x)
            if abs(reqs[i]-reqs[i-1])<abs(reqs[i+1]-reqs[i]):
                x=reqs[i-1]
            else :
                x=reqs[i+1]
        else:
            x=min(reqs)

        reqs.remove(r[-1])
    return r

sstf(requests,15)
    
     
    
    
```

```python
requests=[15,14,11,16,4,35,40]
def look(reqs,start,direction):
    r=[]
    reqs.sort()
    if start not in reqs :
        reqs.append(start)
        reqs.sort()
    if direction.upper()=="LEFT":
        if reqs.index(start)==0:

            r=reqs

        else:

            r=reqs[reqs.index(start):]+reqs[reqs.index(start)-1::-1]

    elif direction.upper()=="RIGHT":
        r=reqs[reqs.index(start)::-1]+reqs[reqs.index(start)+1:]
    else:
        print("enter correct direction")
    return r
look(requests,15,"left")   
```

```python
requests=[15,14,11,16,4,35,40]
def scane(reqs,start,direction,numTracks):
    r=[]
    reqs.sort()

    if start not in reqs :
        reqs.append(start)
        reqs.sort()
    

    if direction.upper()=="LEFT":
        if reqs.index(start)==0:

            r=reqs

        else:

            r=reqs[reqs.index(start):]+[numTracks]+reqs[reqs.index(start)-1::-1]

    elif direction.upper()=="RIGHT":
        r=reqs[reqs.index(start)::-1]+[0]+reqs[reqs.index(start)+1:]
    else:
        print("enter correct direction")
    return r
scane(requests,15,"right",50) 
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter.scrolledtext import ScrolledText
import plotly.graph_objects as go

import plotly.io as pio
import os

class PlotApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Disk Scheduling Algorithms")

        self.requests_label = ttk.Label(master, text="Enter Requests (comma-separated):")
        self.requests_label.grid(row=0, column=0, padx=10, pady=5)
        self.requests_entry = ttk.Entry(master, width=50)
        self.requests_entry.grid(row=0, column=1, padx=10, pady=5)

        self.start_label = ttk.Label(master, text="Enter Starting Position:")
        self.start_label.grid(row=1, column=0, padx=10, pady=5)
        self.start_entry = ttk.Entry(master)
        self.start_entry.grid(row=1, column=1, padx=10, pady=5)

        self.direction_label = ttk.Label(master, text="Enter Direction (LEFT or RIGHT):")
        self.direction_label.grid(row=2, column=0, padx=10, pady=5)
        self.direction_entry = ttk.Entry(master)
        self.direction_entry.insert(tk.END, "LEFT")
        self.direction_entry.grid(row=2, column=1, padx=10, pady=5)

        self.num_tracks_label = ttk.Label(master, text="Enter Number of Tracks:")
        self.num_tracks_label.grid(row=3, column=0, padx=10, pady=5)
        self.num_tracks_entry = ttk.Entry(master)
        self.num_tracks_entry.grid(row=3, column=1, padx=10, pady=5)

        self.unit_time_label = ttk.Label(master, text="Enter Unit Time Required:")
        self.unit_time_label.grid(row=4, column=0, padx=10, pady=5)
        self.unit_time_entry = ttk.Entry(master)
        self.unit_time_entry.grid(row=4, column=1, padx=10, pady=5)

        self.algorithm_label = ttk.Label(master, text="Select Algorithm:")
        self.algorithm_label.grid(row=5, column=0, padx=10, pady=5)
        self.algorithm_combobox = ttk.Combobox(master, values=["FCFS", "SSTF", "SCAN", "LOOK", "CLOOK", "CSCAN"])
        self.algorithm_combobox.grid(row=5, column=1, padx=10, pady=5)

        self.plot_button = ttk.Button(master, text="Plot", command=self.plot_result)
        self.plot_button.grid(row=6, columnspan=2, padx=10, pady=10)

        self.output_text = ScrolledText(master, height=8, width=70, wrap=tk.WORD)
        self.output_text.grid(row=7, columnspan=2, padx=10, pady=10)

        self.table_frame = ttk.Frame(master)
        self.table_frame.grid(row=8, columnspan=2, padx=10, pady=10)

        self.table_label = ttk.Label(self.table_frame, text="Results:")
        self.table_label.grid(row=0, column=0, padx=10, pady=5)

        self.table = ttk.Treeview(self.table_frame, columns=("Path Head", "Tracks Travelled"), show="headings", height=8)
        self.table.heading("Path Head", text="Path Head")
        self.table.heading("Tracks Travelled", text="Tracks Travelled")
        self.table.grid(row=1, column=0, padx=10, pady=5)

    def plot_result(self):
        algorithm = self.algorithm_combobox.get()
        requests = list(map(int, self.requests_entry.get().split(',')))
        start = int(self.start_entry.get())
        direction = self.direction_entry.get().upper()
        num_tracks = int(self.num_tracks_entry.get())
        unit_time = float(self.unit_time_entry.get())

        if algorithm == "FCFS":
            result = fcfs(requests, start)
        else:
            if algorithm == "SSTF":
                result = sstf(requests, start)
            elif algorithm == "SCAN":
                result = scane(requests, start, direction)
            elif algorithm == "LOOK":
                result = look(requests, start, direction)
            elif algorithm == "CLOOK":
                result = clook(requests, start, direction, num_tracks)
            elif algorithm == "CSCAN":
                result = cscane(requests, start, direction, num_tracks)
            else:
                return

        differences = diff(result)
        total_tracks = sum(differences)
        time_to_travel_tracks = total_tracks * unit_time

        fig = go.Figure()
        
        fig.add_trace(go.Scatter(x=result, y=list(range(1, len(result) + 1)), mode='markers+lines',
                                 marker=dict(size=10, color='yellow', symbol='circle-open')))



        for i, diff_value in enumerate(differences):
            fig.add_annotation(x=(result[i] + result[i + 1]) / 2, y=(i + 1 + i + 2) / 2,
                               text=f"Difference: {diff_value}",
                               showarrow=False,
                               font=dict(size=12, color="green"))
        fig.update_layout(title=f"Disk Scheduling ({algorithm})", xaxis_title="Tracks", yaxis_title="Request Order")

        # Update output text
        self.output_text.delete('1.0', tk.END)
        self.output_text.insert(tk.END, f"Total number of tracks traveled is {total_tracks}\n")
        self.output_text.insert(tk.END, f"Time required to travel 1 track is {unit_time} unit time\n")
        self.output_text.insert(tk.END, f"Time required to travel {total_tracks} tracks is {time_to_travel_tracks} unit time\n")

        # Update table
        self.update_table(result, differences)

        # Save the Plotly figure to an HTML file
        plot_file = "plot.html"
        pio.write_html(fig, plot_file)

        # Open the HTML file in a web browser
        os.system(plot_file)

    def update_table(self, result, differences):
        self.table.delete(*self.table.get_children())
        for i, (track, diff_value) in enumerate(zip(result[:-1], differences)):
            self.table.insert("", "end", values=(f"{track} - {result[i + 1]}", diff_value))


def fcfs(reqs, start):
    if start not in reqs :
        reqs.insert(0,start)
    else:
        reqs.remove(start)
        reqs.insert(0,start)
    return reqs

def sstf(reqs, start):
    if start not in reqs:
        reqs.append(start)
    x = start
    r = []
    while reqs:
        r.append(x)
        reqs.sort()

        if len(reqs) >= 2:
            i = reqs.index(x)
            if abs(reqs[i] - reqs[i - 1]) < abs(reqs[i + 1] - reqs[i]):
                x = reqs[i - 1]
            else:
                x = reqs[i + 1]
        else:
            x = min(reqs)

        reqs.remove(r[-1])
    return r

def scane(reqs, start, direction):
    r = []
    reqs.sort()

    if start not in reqs:
        reqs.append(start)
        reqs.sort()

    if direction.upper() == "LEFT":
        if reqs.index(start) == 0:
            r = reqs
        else:
            r = reqs[reqs.index(start):] + [num_tracks] + reqs[reqs.index(start) - 1::-1]
    elif direction.upper() == "RIGHT":
        r = reqs[reqs.index(start)::-1] + [0] + reqs[reqs.index(start) + 1:]
    else:
        print("Enter correct direction")
    return r

def clook(reqs, start, direction, num_tracks):
    r = []
    reqs.sort()

    if start not in reqs:
        reqs.append(start)
        reqs.sort()

    if direction.upper() == "LEFT":
        if reqs.index(start) == 0:
            r = reqs
        else:
            r = reqs[reqs.index(start):] + reqs[:reqs.index(start)]
    elif direction.upper() == "RIGHT":
        r = reqs[reqs.index(start)::-1] + reqs[:reqs.index(start):-1]
    else:
        print("Enter correct direction")
    return r

def cscane(reqs, start, direction, num_tracks):
    r = []
    reqs.sort()

    if start not in reqs:
        reqs.append(start)
        reqs.sort()

    if direction.upper() == "LEFT":
        if reqs.index(start) == 0:
            r = reqs
        else:
            r = reqs[reqs.index(start):] + [num_tracks] + [0] + reqs[:reqs.index(start)]
    elif direction.upper() == "RIGHT":
        r = reqs[reqs.index(start)::-1] + [0] + [num_tracks] + reqs[:reqs.index(start):-1]
    else:
        print("Enter correct direction")
    return r
def look(reqs,start,direction):
    r=[]
    reqs.sort()
    if start not in reqs :
        reqs.append(start)
        reqs.sort()
    if direction.upper()=="LEFT":
        if reqs.index(start)==0:

            r=reqs

        else:

            r=reqs[reqs.index(start):]+reqs[reqs.index(start)-1::-1]

    elif direction.upper()=="RIGHT":
        r=reqs[reqs.index(start)::-1]+reqs[reqs.index(start)+1:]
    else:
        print("enter correct direction")
    return r

def diff(x):
    return [abs(x[i + 1] - x[i]) for i in range(len(x) - 1)]

def main():
    root = tk.Tk()
    app = PlotApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
